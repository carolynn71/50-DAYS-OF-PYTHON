```python


# Python Operators
# Operators are special symbols or keywords that perform operations on one or more operands.
# Operands are the values that operators act upon.

# 1. Arithmetic Operators:
# Used to perform mathematical operations like addition, subtraction, multiplication, division, etc.
# +	Addition	x + y
# -	Subtraction	x - y
# *	Multiplication	x * y
# /	Division	x / y (float division)
# %	Modulus		x % y (remainder)
# **	Exponentiation	x ** y (x to the power of y)
# //	Floor division	x // y (integer division)

print("\n--- 1. Arithmetic Operators ---")
a = 10
b = 3
print(f"a + b = {a + b}") # 13
print(f"a - b = {a - b}") # 7
print(f"a * b = {a * b}") # 30
print(f"a / b = {a / b}") # 3.333...
print(f"a % b = {a % b}") # 1
print(f"a ** b = {a ** b}") # 1000
print(f"a // b = {a // b}") # 3

# 2. Assignment Operators:
# Used to assign values to variables.
# =	Assigns value from right side to left side operand	c = a + b assigns value of a + b into c
# +=	Add AND assignment	c += a is equivalent to c = c + a
# -=	Subtract AND assignment	c -= a is equivalent to c = c - a
# *=	Multiply AND assignment	c *= a is equivalent to c = c * a
# /=	Divide AND assignment	c /= a is equivalent to c = c / a
# %=	Modulus AND assignment	c %= a is equivalent to c = c % a
# **=	Exponent AND assignment	c **= a is equivalent to c = c ** a
# //=	Floor-divide AND assignment	c //= a is equivalent to c = c // a

print("\n--- 2. Assignment Operators ---")
x = 5
print(f"Initial x = {x}") # 5
x += 3 # x = x + 3
print(f"x after x += 3: {x}") # 8
x -= 2 # x = x - 2
print(f"x after x -= 2: {x}") # 6
x *= 4 # x = x * 4
print(f"x after x *= 4: {x}") # 24
x /= 6 # x = x / 6
print(f"x after x /= 6: {x}") # 4.0

# 3. Comparison (Relational) Operators:
# Used to compare two values and return a boolean (True or False).
# ==	Equal to		x == y
# !=	Not equal to		x != y
# >	Greater than		x > y
# <	Less than		x < y
# >=	Greater than or equal to	x >= y
# <=	Less than or equal to		x <= y

print("\n--- 3. Comparison Operators ---")
p = 7
q = 10
print(f"p == q: {p == q}") # False
print(f"p != q: {p != q}") # True
print(f"p > q: {p > q}") # False
print(f"p < q: {p < q}") # True
print(f"p >= 7: {p >= 7}") # True
print(f"q <= 10: {q <= 10}") # True

# 4. Logical Operators:
# Used to combine conditional statements.
# and	Logical AND	Returns True if both statements are true	x < 5 and x < 10
# or	Logical OR	Returns True if one of the statements is true	x < 5 or x < 4
# not	Logical NOT	Reverses the result, returns False if the result is true	not(x < 5 and x < 10)

print("\n--- 4. Logical Operators ---")
r = 6
print(f"r > 5 and r < 10: {r > 5 and r < 10}") # True
print(f"r < 5 or r < 10: {r < 5 or r < 10}") # True
print(f"not(r < 5 and r < 10): {not(r < 5 and r < 10)}") # True

# 5. Identity Operators:
# Used to compare the objects, not if they are equal, but if they are actually the same object, with the same memory location.
# is	Returns True if both variables are the same object	x is y
# is not	Returns True if both variables are not the same object	x is not y

print("\n--- 5. Identity Operators ---")
l1 = [1, 2, 3]
l2 = [1, 2, 3]
l3 = l1
print(f"l1 is l3: {l1 is l3}") # True (l1 and l3 refer to the same object)
print(f"l1 is l2: {l1 is l2}") # False (l1 and l2 are different objects, even if content is same)
print(f"l1 == l2: {l1 == l2}") # True (l1 and l2 have the same content)
print(f"l1 is not l2: {l1 is not l2}") # True

# 6. Membership Operators:
# Used to test if a sequence is presented in an object.
# in	Returns True if a sequence with the specified value is present in the object	x in y
# not in	Returns True if a sequence with the specified value is not present in the object	x not in y

print("\n--- 6. Membership Operators ---")
my_list = [10, 20, 30, 40]
print(f"20 in my_list: {20 in my_list}") # True
print(f"50 in my_list: {50 in my_list}") # False
print(f"50 not in my_list: {50 not in my_list}") # True

my_string = "Hello World"
print(f"'World' in my_string: {'World' in my_string}") # True
print(f"'Python' not in my_string: {'Python' not in my_string}") # True

# 7. Bitwise Operators (Advanced - operate on bits):
# &	AND	Sets each bit to 1 if both bits are 1
# |	OR	Sets each bit to 1 if one of two bits is 1
# ^	XOR	Sets each bit to 1 if only one of two bits is 1
# ~	NOT	Inverts all the bits
# <<	Zero fill left shift	Shift left by pushing zeros in from the right and let the leftmost bits fall off
# >>	Signed right shift	Shift right by pushing copies of the leftmost bit in from the left, and let the rightmost bits fall off

print("\n--- 7. Bitwise Operators (Binary level) ---")
a_bit = 10  # Binary: 1010
b_bit = 4   # Binary: 0100
print(f"a_bit & b_bit (10 & 4): {a_bit & b_bit}") # 0000 -> 0
print(f"a_bit | b_bit (10 | 4): {a_bit | b_bit}") # 1110 -> 14
print(f"a_bit ^ b_bit (10 ^ 4): {a_bit ^ b_bit}") # 1110 -> 14
print(f"~a_bit (~10): {~a_bit}") # -11 (depends on system's representation of negative numbers)
print(f"a_bit << 2 (10 << 2): {a_bit << 2}") # 101000 -> 40 (equivalent to a_bit * 2**2)
print(f"a_bit >> 2 (10 >> 2): {a_bit >> 2}") # 0010 -> 2 (equivalent to a_bit // 2**2)
```
