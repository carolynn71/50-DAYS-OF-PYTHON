```python
#LITERALS
```

```python
#literals is a raw data given in variable. In python, there are various types of literals they are as follows:-
#NUMERIC
#STRING
#BOOLEAN
#SPECIAL
```

```python
#1. NUMERICAL
a=100  #-->decimal
print(a)
c=0o310  #-->octal
print(c)
b=0b1010  #-->binary
print(b)
d=0x12c   #-->Hexadecimal
print(d)

#FLOAT LITERAL
float_1=100.10
float_2=1.5e2
print(float_1,float_2)

#COMPLEX LITERAL
x=3.14j
print(x,x.imag,x.real)
```

```python

#2. STRING
string='This is python'
strings="THIS IS PYTHON"
char='c'
multiline_str="""THIS IS A MULTILINE STRING WITH MORE
THAN ONE LINE OF CODE"""
print(string)
print(strings)
print(char)
print(multiline_str)
```

```python
#3. BOOLEAN
a=True+4
b=False+10
print(a)
print(b)
```

```python

#4. SPECIAL
#this is variable declaration for null.
a=None
print(a)
```
